##### This is an example of using imageio to display and image
###### First, I had to pip install imageio. Then I imported 2 libraries. The following code is displaying my image 

```python
import imageio
import matplotlib.pyplot as plt

roses = imageio.imread("Rainbow Roses.jpg")
plt.imshow(roses)
plt.axis("off")
plt.show()
```

##### This is an example of converting the file format on my image

```python
input_file = 'Rainbow Roses.jpg'
output_file = 'Rainbow Roses.png'

image = imageio.imread(input_file)
imageio.imwrite(output_file, image, format='PNG')
```
